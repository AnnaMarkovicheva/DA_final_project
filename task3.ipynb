# Проект 1

## Задание 3

*В игре Plants & Gardens каждый месяц проводятся тематические события, ограниченные по времени. В них игроки могут получить уникальные предметы для сада и персонажей, дополнительные монеты или бонусы. Для получения награды требуется пройти ряд уровней за определенное время. С помощью каких метрик можно оценить результаты последнего прошедшего события?*

1) В первую очередь, я бы рассмотрела метрику daily active users - как она менялась по дням на фоне проведения тематического события, были ли им привлечены игроки. В добавление - не "проседала" ли эта метрика через какое-то количество дней после начала события (то есть, было ли оно способно удержать пользователя, вызвало ли у него продолжительный (более одного дня) интерес)?

2) Также я бы взглянула на процент игроков, которые решили принять участие в событии (назовём эту метрику "конверсия в событие"), чтобы посмотреть, связаны ли изменения DAU с событием. Например, если критерий DAU повысился, а процент пользователей, вовлеченных в событие, крайне мал, то может, что-то другое повлияло на рост DAU (и в таком случае, стоит ли говорить об успешности события?)

3) Стоит иметь в виду и прогресс игроков по уровням. Сколько игроков прошли последний уровень и получили награду из всех тех, кто принял участие в событии? Был ли уровень (или уровни), на котором "отваливалось" внушительное количество игроков? Если да, то можно рассмотреть, с чем это связано: баг, мешающий пройти уровень; специальное препятствие, чтобы не все игроки могли легко получить награду; или фича (другое)?

4) Дополнительно можно проверить такие метрики, как ARPU, ARPPU, конверсия в платящего пользователя - как менялся доход во время проведения события. А также, регистрировались ли новые пользователи у игры (но здесь есть риск, что эта метрика может быть не связана с событием, поскольку иначе свежеиспеченные игроки бы в него зашли с большой вероятностью, а не просто стали играть).

*Предположим, в другом событии мы усложнили механику событий так, что при каждой неудачной попытке выполнения уровня игрок будет откатываться на несколько уровней назад. Изменится ли набор метрик оценки результата? Если да, то как?*

Я бы оставила все те же метрики, но больше бы сосредоточилась на churn rate / оттоке пользователей. Конечно, с одной стороны, откаты на уровни назад направлены на то, чтобы целеустремленный игрок проводил в игре больше времени (или вкладывал деньги в игру, чтобы получить награду). С другой стороны, - если есть, например, несбалансированные уровни, то весьма вероятно, что откат назад будет фрустрировать игрока, что может вызвать у него сильное нежелание продолжать участие в событии, если не в игре в целом. Иными словами, негативные эмоции приводят к смещению акцента на негативную метрику churn (в отличие от более позитивного критерия retention rate - сколько пользователей вернулись =)
